# FEATURE ENGINEERING

```python
import numpy as np

# Create correlation matrix
corr_matrix = df.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find features with correlation greater than 0.8
to_drop = [column for column in upper.columns if any(upper[column] > 0.8)]

# Drop features
df.drop(to_drop, axis=1, inplace=True)
```

#### As we aready know that there should be no value named unknown in the target variable as it shows no value we need to remove the rows containing the target variable named 'UNKNOWN'

```python
x=df.drop(['SIM_TYPE_NAME','SUBSURVEY'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
ly=LabelEncoder()
y=ly.fit_transform(y)
```

```python
y
```

```python
ly.classes_
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

###We can see that there is class imbalance hence we need to perform under /over sampling

```python
!pip install imblearn
```

```python
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
```

```python
rus=RandomUnderSampler()
rus_x_train,rus_y_train=rus.fit_resample(x_train,y_train)
```

```python
print(Counter(rus_y_train))
```

```python
import xgboost as xgb
xgb_model=xgb.XGBClassifier(random_state=0)
xgb_model.fit(rus_x_train,rus_y_train)
```

```python
yp=xgb_model.predict(x_test)
```

```python
 from sklearn.metrics import confusion_matrix,accuracy_score
 cm=confusion_matrix(y_test,yp)
 cm
```

```python
accuracy_score(y_test,yp)
```
